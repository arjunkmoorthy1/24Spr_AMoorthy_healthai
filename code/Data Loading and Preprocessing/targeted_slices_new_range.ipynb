```python
import os
import numpy as np
import pandas as pd
import pydicom
import shutil

# Assuming dicom_dir_path and targeted_bx_path are defined earlier in your code
output_dir_path = os.path.join(dicom_dir_path, 'targeted_bx_tract_slices')

# Ensure the output directory exists
os.makedirs(output_dir_path, exist_ok=True)

# Load the Excel file
df = pd.read_excel(targeted_bx_path)

def find_and_copy_slices_in_range(patient_id, z_start, z_end, biopsy_index, primary_gleason):
    patient_dicom_dir = os.path.join(dicom_dir_path, patient_id)
    copied_slices_count = 0

    # Ensure correct order of start and end for z-coordinates
    z_min, z_max = sorted([z_start, z_end])

    # Determine folder name based on presence of Primary Gleason score
    cancer_status = "cancer" if pd.notnull(primary_gleason) else "non_cancer"
    folder_name = f"biopsy_{biopsy_index}_{cancer_status}"

    # Define output directory for this specific biopsy
    output_biopsy_dir = os.path.join(output_dir_path, patient_id, folder_name)
    os.makedirs(output_biopsy_dir, exist_ok=True)

    # Walk through all subdirectories
    for root, dirs, files in os.walk(patient_dicom_dir):
        for file in files:
            if file.endswith(".dcm"):
                file_path = os.path.join(root, file)
                try:
                    ds = pydicom.dcmread(file_path)
                    # Check for the existence of 'ImagePositionPatient' in the dataset
                    if 'ImagePositionPatient' in ds:
                        dicom_z_coord = ds.ImagePositionPatient[2]
                        if z_min <= dicom_z_coord <= z_max:
                            shutil.copy(file_path, output_biopsy_dir)
                            copied_slices_count += 1
                    else:
                        # Skip this file and continue to the next
                        continue
                except Exception as e:
                    print(f"Error processing file {file_path}: {e}")
                    # Optionally, you could log or handle specific exceptions

    print(f"Copied {copied_slices_count} slices for biopsy {biopsy_index} ({cancer_status}) of patient {patient_id} to {output_biopsy_dir}")
```

```python
import os
import shutil

def remove_duplicate_tracts(patient_dir_base, new_patient_dir_base):
    os.makedirs(new_patient_dir_base, exist_ok=True)  # Ensure new base directory exists

    # Iterate over each patient directory
    for patient_id in os.listdir(patient_dir_base):
        patient_path = os.path.join(patient_dir_base, patient_id)
        if not os.path.isdir(patient_path):
            continue  # Only process directories

        new_patient_dir = os.path.join(new_patient_dir_base, patient_id)
        os.makedirs(new_patient_dir, exist_ok=True)  # Create new patient directory

        # Reset tracking for each patient
        cancer_files = set()

        # Collect all cancer and non_cancer tract paths for the patient
        cancer_tracts, non_cancer_tracts = {}, {}
        for tract_name in os.listdir(patient_path):
            tract_path = os.path.join(patient_path, tract_name)
            if os.path.isdir(tract_path):
                if 'non_cancer' in tract_name:
                    non_cancer_tracts[tract_name] = tract_path
                else:  # Assuming other tracts are cancer by default
                    cancer_tracts[tract_name] = tract_path

        # Process cancer tracts: copy all files and record filenames
        for tract_name, tract_path in cancer_tracts.items():
            new_tract_dir = os.path.join(new_patient_dir, tract_name)
            os.makedirs(new_tract_dir, exist_ok=True)
            for file in os.listdir(tract_path):
                if file.endswith('.dcm'):
                    cancer_files.add(file)  # Track file names within this patient
                    shutil.copy2(os.path.join(tract_path, file), os.path.join(new_tract_dir, file))

        # Process non_cancer tracts: copy files only if they're not found in cancer_files
        for tract_name, tract_path in non_cancer_tracts.items():
            # Identify non-overlapping files for this non_cancer tract
            files_to_copy = [file for file in os.listdir(tract_path) if file.endswith('.dcm') and file not in cancer_files]

            # If there are files to copy, meaning no overlap, proceed to copy them
            if files_to_copy:
                new_tract_dir = os.path.join(new_patient_dir, tract_name)
                os.makedirs(new_tract_dir, exist_ok=True)
                for file in files_to_copy:
                    shutil.copy2(os.path.join(tract_path, file), os.path.join(new_tract_dir, file))
```

```python
# Example usage
patient_dir_base = '/Users/arjunmoorthy/Desktop/t2axialimages/targeted_bx_tract_slices'
new_patient_dir_base = '/Users/arjunmoorthy/Desktop/t2axialimages/after_duplicate_tracts_removed'
remove_duplicate_tracts(patient_dir_base, new_patient_dir_base)
```

```python

# Base directory where the patient biopsy tracts are stored
base_dir = '/Users/arjunmoorthy/Desktop/t2axialimages/after_duplicate_tracts_removed'

# Destination directories for categorized data
dest_dir_cancer = '/Users/arjunmoorthy/Desktop/Research_Capstone/ImageData/final_train_data/cancer_tracts'
dest_dir_non_cancer = '/Users/arjunmoorthy/Desktop/Research_Capstone/ImageData/final_train_data/non_cancer_tracts'

# Create destination directories if they don't exist
os.makedirs(dest_dir_cancer, exist_ok=True)
os.makedirs(dest_dir_non_cancer, exist_ok=True)

# Loop through all directories in the base directory
for root, dirs, files in os.walk(base_dir):
    for dir_name in dirs:
        current_path = os.path.join(root, dir_name)
        
        # Check if the folder name ends with '_non_cancer'
        if dir_name.endswith('_non_cancer'):
            dest_path = os.path.join(dest_dir_non_cancer, dir_name)
        # Check if the folder name ends with '_cancer'
        elif dir_name.endswith('_cancer'):
            dest_path = os.path.join(dest_dir_cancer, dir_name)
        else:
            continue  # Skip directories that do not match the pattern

        # Check if the destination path exists, create if not
        if not os.path.exists(dest_path):
            os.makedirs(dest_path, exist_ok=True)
        
        # Copy all files from current directory to the appropriate destination directory
        for file in os.listdir(current_path):
            src_file = os.path.join(current_path, file)
            dest_file = os.path.join(dest_path, file)
            shutil.copy(src_file, dest_file)

        print(f"Files from {current_path} copied to {dest_path}")

# Confirm completion
print("Data categorization complete.")
```
